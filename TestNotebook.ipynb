```python
import pandas as pd
from glob import glob
import gp, gp.data
import re

#numFiles carries CSV files
numFiles = []
all_paths = glob("Untitled Folder/*.csv")
all_paths

#df_preList carries dataframes
df_preList = []
score_df = pd.DataFrame([])

#this for loop will fill up the list numFiles with all the CSV files
for filename in all_paths:
    numFiles.append(filename)
    
# this for loop will take every CSV in the list numFiles and create their corresponding dataframe
# these dataframes are then added to the list df_preList
for i in range(len(all_paths)):
        df_csv = pd.read_csv(numFiles[i])
        df_preList.append(df_csv)
        
#this for loop will extract the "score" column from every csv dataframe and 
#add them to an existing one that only carries name, moa, and the first score
counter = 0
for i in range(len(all_paths)):
    if counter == 0:
        df0 = df_preList[counter]
        df0 = df0.set_index("drug")
        #df0.rename(index={'drug':'name'})
        df0.index.names = ['NAME']
        tempstring = str (numFiles[i])
        tempstring =(tempstring[16:31:1])
        
        neworder = ['moa','score']
        df0=df0[neworder]
        
        df0=df0.rename(columns = {'score':tempstring,'moa':'Description'})
        
    else:
        tempstring = str (numFiles[i])
        tempstring =(tempstring[16:31:1])
        col_name = tempstring
        temp = df_preList[i]
        temp = temp.set_index("drug")
        df0[col_name] = temp["score"]
        
    counter += 1
    
def standarize_string(what):
    ascii_8bit = re.sub(r'[^ -~]', '', what).lower() # keeping only 8-bit ascii, making lowercase (http://www.catonmat.net/blog/my-favorite-regex/)
    return ascii_8bit
    #return re.sub(r'[ _-]', '', ascii_8bit) # remove space, underscore, and dash (" _-")
    
for index, row in df0.iterrows():
    #moa_column = str (row['Description'])
#     row['Description'] = standarize_string(str (row['Description']))
    df0.loc[index,'Description'] = standarize_string(str (row['Description']))
#     print(row['Description'])
    #moa_column.encode('ascii', 'ignore')
    
#this block of code will run through every entry under the description column
#and will check if the contents are in ascii
#if not, those unrecognized characters will be converted to their ascii equivalents
```

```python
gpserver = gp.GPServer('https://cloud.genepattern.org/gp','gabrield706', 'b6bekrien')
gpserver.get_task_list()
```

```python
module = gp.GPTask(gpserver, "DiSCoVER")  # Obtaining GPTask by module name
#module = gp.GPTask(gpserver, "urn:lsid:broad.mit.edu:cancer.software.genepattern.module.analysis:00020:5")  # Obtaining GPTask by LSID
module.param_load()
```

```python
params_list = module.get_parameters()  # Get the list of GPParam objects

for param in params_list:  # Loop through each parameter
    print( param.get_name() )  # Print the parameter's name
    print( param.get_description() )  # Print the parameter's description
    print( param.get_default_value() )  # Print the parameter's default value
    print( param.is_optional() )  # Print whether the parameter is optional
```

```python
job_spec = module.make_job_spec()  # Create the GPJobSpec
uploaded_gene_file = gpserver.upload_file("TCGA-A7-A0CE-01", "TCGA-A7-A0CE-01.csv")  # Upload the input file
job_spec.set_parameter("gene_expression", uploaded_gene_file.get_url())  # Attach the input file to the correct parameter
for param in module.get_parameters():  # Loop through all the other parameters and set their default values
    if param.get_name() == "use_custom_control" and param.get_default_value() == False:
        job_spec.set_parameter(param.get_name(), True)
    
    elif param.get_name() == "control":
        uploaded_cc_file = gpserver.upload_file("TCGA-A7-A0CE-11", "TCGA-A7-A0CE-11.csv")
        job_spec.set_parameter("use_custom_control", uploaded_cc_file.get_url())
    
    elif param.get_name() != "gene_expression" and param.get_default_value() != None:  # If it's not the parameter we just set, and if it has a default value
        job_spec.set_parameter( param.get_name(), param.get_default_value() )  # Set the default value
    #elif param.get_name() == "use_custom_control" and param.get_default_value() == False:
        #job_spec.set_parameter(param.get_name(), True)
        #gpserver.upload_file("TCGA-A7-A0CE-11", "TCGA-A7-A0CE-11.csv")
        #job_spec.set_parameter("use_custom_control", uploaded_cc_file.get_url())
```

```python
#job = gpserver.run_job(job_spec)  # This will halt execution until the job is complete
job = gpserver.run_job(job_spec, False)  # This will return the job object and continue execution even if the job isn't finished
```

```python
job.is_finished()  # Quaries the server and returns True if the job is complete, False otherwise
job.get_info()  # Returns a brief description of the job's current state
print(job.is_finished())
```
