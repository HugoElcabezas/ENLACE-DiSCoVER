```python
import pandas as pd
import os
from matplotlib import pyplot as plt
import numpy as np
```

```python
folder_to_open = 'CSVresults2'
list_of_files = os.listdir(folder_to_open)

out_dataframe = pd.DataFrame()
for current_file in list_of_files:
    if current_file.endswith('.csv'):
        current_sample = current_file.strip('.csv')
        sample_df = pd.read_csv(folder_to_open+'/'+current_file,index_col=0)
#         out_dataframe['drug'] = sample_df['drug']
        out_dataframe[current_sample] = sample_df['score']
raw_df = out_dataframe
# out_dataframe.sort_values(by='TCGA-BH-A0AU-0_DiSCoVER_result')
raw_df
```

```python
raw_df_w = raw_df.rank(method = 'first', ascending = False)
raw_df_x = raw_df_w.reset_index()
raw_df_y = raw_df_x.sort_values(by =['drug'])
raw_df_rankings = raw_df_y.reset_index(drop=True)
raw_df_rankings
```

```python
folder_to_open = 'CSVresults3'
list_of_files = os.listdir(folder_to_open)

out_dataframe = pd.DataFrame()
for current_file in list_of_files:
    if current_file.endswith('.csv'):
        current_sample = current_file.strip('.csv')
        sample_df = pd.read_csv(folder_to_open+'/'+current_file,index_col=0)
#         out_dataframe['drug'] = sample_df['drug']
        out_dataframe[current_sample] = sample_df['score']
fpkm_df = out_dataframe
# out_dataframe.sort_values(by='TCGA-BH-A0AU-0_DiSCoVER_result')
fpkm_df
```

```python
fpkm_df_w = fpkm_df.rank(method = 'first', ascending = False)
fpkm_df_x = fpkm_df_w.reset_index()
fpkm_df_y = fpkm_df_x.sort_values(by =['drug'])
fpkm_df_rankings = fpkm_df_y.reset_index(drop=True)
fpkm_df_rankings
```

```python
folder_to_open = 'CSVresults2'
list_of_files = os.listdir(folder_to_open)
list_of_files.sort()

patient_nameList =[]
ken_valueList = []

for current_file in list_of_files:
    if current_file.endswith('.csv'):
        current_sample = current_file.strip('.csv')
        patient_nameList.append(current_sample)
    else:
        pass
#print(patient_nameList)

for current_file in list_of_files:
    if "ipynb_checkpoints" in current_file:
        continue
    else:
        temp = pd.DataFrame()
        current_sample = current_file.strip('.csv')
        temp[current_sample+"raw"] = raw_df[current_sample]
        temp[current_sample+"fpkm"] = fpkm_df[current_sample]
        #print(temp.head())
        ken_tau = temp.corr(method='kendall')
        ken_number = ken_tau.loc[current_sample+"raw",current_sample+"fpkm"]
        ken_valueList.append(ken_number)
        display(ken_tau)
#display(ken_valueList)

```

```python
plt.bar(patient_nameList, ken_valueList, label = 'Kendall Coefficients')
plt.style.use('seaborn-muted')
plt.xticks(rotation='vertical')
plt.xlabel('Patients', labelpad=20)
plt.ylabel('Kendall Coefficients', labelpad=20)

plt.show()
```
