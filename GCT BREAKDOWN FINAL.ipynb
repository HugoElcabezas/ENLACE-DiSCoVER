```python
import gp, gp.data
import pandas as pd
from glob import glob
import re
import os
import time
#import wget
import getpass
from io import StringIO
import io

df_new = pd.read_csv('https://datasets.genepattern.org/data/TCGA_HTSeq_FPKM/BRCA/BRCA_FPKM.gct',
                  header=2,sep='\t').drop('Description',axis=1)

list_new_names = df_new.drop('Name',axis=1).columns

df_old = pd.read_csv('https://datasets.genepattern.org/data/ccmi_tutorial/2018-09-20/BRCA_HUGO_symbols.preprocessed.gct',
                  header=2,sep='\t').drop('Description',axis=1)

list_old_names = df_old.drop('Name',axis=1).columns

df_old.index.names = ['NAME']
df_old = df_old.sort_values(by=['NAME'])

FinalDataFrame = pd.DataFrame()
FinalDataFrame['Name'] = df_new['Name'].values
FinalDataFrame['Description'] = df_new['Name'].values

FinalDataFrame = FinalDataFrame.set_index("Name")
FinalDataFrame.index.names = ['NAME']

for name in list_old_names:
    FinalDataFrame[name] = df_new[name].values
    

    
def standarize_string(what):
    ascii_8bit = re.sub(r'[^ -~]', '', what).lower() # keeping only 8-bit ascii, making lowercase (http://www.catonmat.net/blog/my-favorite-regex/)
    return ascii_8bit
    #return re.sub(r'[ _-]', '', ascii_8bit) # remove space, underscore, and dash (" _-")
    
for index, row in FinalDataFrame.iterrows():
    
    FinalDataFrame.loc[index,'Description'] = standarize_string(str (row['Description'])) 

FinalDataFrame['Description']='na'
FinalDataFrame.sort_values(by=['NAME'])
    
    
    
    
FinalDataFrame


```

```python
output_name = "df_brca_fpkm"
gp.data.write_gct(FinalDataFrame, file_path='./'+output_name)
print("DataFrame Successfully Created! Please look for" +output_name+ "in your home directory")
```
